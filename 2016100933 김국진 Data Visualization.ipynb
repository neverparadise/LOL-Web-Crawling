```python
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
from os import listdir
from os.path import isfile, join




class Visualization:
    def __init__(self, given_nickname):
        self.nickname = given_nickname
        self.filepath = ""
        self.df = ""
    
    def setFilePath(self):
        self.filepath = "./data/" + self.nickname + ".csv"
        print("filepath : " + self.filepath)
       
    def read_csv(self):
        self.df = pd.read_csv(self.filepath, delimiter=',', encoding='CP949', engine='python') #engine='python' 을 설정하지 않으면 읽기 오류가 발생
        
        print("raw data")
        print(self.df)
        return self.df
    def process_csv(self):
        self.df = self.df.drop(self.df.columns[0], axis='columns')   #인덱스가 포함된 열을 제거해줍니다.
        print("processed data")
        return self.df

files = [f for f in listdir('./data') if isfile(join('./data', f))]
print(files)    

name = input("데이터를 분석하고 싶은 사용자의 닉네임을 입력하세요 : ")
User = Visualization(name)
User.setFilePath()
User.read_csv()
User.process_csv()
```

```python
ax = User.df.plot(kind='bar', title ='Game Record per Day', figsize=(20, 10), legend=True, fontsize = 12)
ax.set_xlabel('GameDate', fontsize=12)
ax.set_ylabel('Number', fontsize=12)
ax.set_yticks(np.arange(1, 30, 1.0))
ax.legend(['GameLength', 'Frequency'], fontsize=12) 
```

```python

FreCnt = User.df['Frequency'].value_counts()
#print(type(FreCnt))
sum = 0
for elem in FreCnt:
    sum += elem

prob = FreCnt[1]/sum #게임을 6개월 중에서 한 판만 즐긴 확률


print(User.nickname + " 님이 최근 6개월간 게임을 한 판만 즐긴 횟수는 전체 " + str(sum) +"일 중 " + str(FreCnt[1]) + "번입니다...")
print(User.nickname + " 님이 최근 6개월간 게임을 한 판만 즐긴 비율은 " + str(prob) + "입니다...")
    
```
